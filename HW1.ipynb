```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import seaborn as sns
from termcolor import colored
```

### A List Of All The Possible Data Types

```python
# array
data = pd.read_csv("iris.csv")
data1 = data.to_numpy()

# dataframe 
data2 = pd.read_csv("iris.csv")

# str & csv
data3="iris.csv"

# str & json
data4="iris.json"

# no type (random)
data5 = []
```

```python
class DataOperations:
    
    def __init__(self, data):
        """Helper function for changing any data type into a dataframe."""
        data_type = type(data) 
        
        if data_type == np.ndarray:
            df = pd.DataFrame(data, columns=['sepalLength','sepalWidth','petalLength','petalWidth','species'])
            self.df = df.infer_objects()              # Converts Object to Float 
            print(colored('\033[1m'+'An array was converted into a dataframe.', 'blue'))
                    
        elif data_type == pd.core.frame.DataFrame:
            self.df = data
            print(colored('\033[1m'+'You already have a dataframe.', 'blue'))
            
        elif data_type == str and data.endswith('.csv'):
            self.df=pd.read_csv(data)
            print(colored('\033[1m'+'A dataframe was created using a CSV file.', 'blue'))
           
        elif data_type == str and data.endswith('.json'):
            self.df = pd.read_json(data)
            print(colored('\033[1m'+'A dataframe was created using a JSON file.', 'blue'))
            
        else:
            self.df = pd.DataFrame(np.random.randint(0,100,size=(150, 4)), columns=list('ABCD'))
            print(colored('\033[1m'+'A dataframe with random numbers was generated.', 'blue'))
    
    def summary_stats(self): 
        """Provides the summary statistics of the data."""
        print(self.df.info(), self.df.describe())
            
    def visualizer(self):
        """Helper function for data visualization."""
        a4_dims = (10, 8)
        fig, axs = plt.subplots(figsize=a4_dims,ncols=3)
        sns.violinplot(x=self.df.iloc[:,0], y=self.df.iloc[:,1], data=self.df, ax=axs[0])
        sns.scatterplot(x=self.df.iloc[:,2], y=self.df.iloc[:,3], data=self.df, ax=axs[1])
        sns.countplot(x=self.df.iloc[:,2], data=self.df)
        

DataOperations_array = DataOperations(data1)
DataOperations_array.summary_stats()
DataOperations_array.visualizer()
print("\n")

DataOperations_df = DataOperations(data2)
DataOperations_df.summary_stats()
DataOperations_df.visualizer()
print("\n")

DataOperations_str_csv = DataOperations(data3)
DataOperations_str_csv.summary_stats()
DataOperations_str_csv.visualizer()
print("\n")

DataOperations_str_json = DataOperations(data4)
DataOperations_str_json.summary_stats()
DataOperations_str_json.visualizer()
print("\n")

DataOperations_random = DataOperations(data5)
DataOperations_random.summary_stats()
DataOperations_random.visualizer()
```
